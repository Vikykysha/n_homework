```python
def mk_dfrm_from_dict_col(dataframe, col_dict_name, *columns):
    '''Делает новый датафрейм из столбца датафрейма типа объект'''
    src_lst = []
    for index, cols in dataframe.iterrows():
        for d_row in eval(cols[col_dict_name]):
            col_n = [d_row[x] for x in  columns]
            src_lst.append((index, *col_n))
    return pd.DataFrame(src_lst)
```

```python
df_genres = pd.merge(df,genres,how = 'inner', left_on = 'id', right_on = 'movieid')
```

Сравниваем количество релизова по жанрам, для этого использовали seaborn и график countplot, переменные категориальные

```python

sns.countplot(df_genres[df_genres['genre'].isin(['Comedy','Drama','Crime','Romance'])]['genre'])
```

```python
#Выберем топ-5 самых выпускаемых за все время жанров
df1 = pd.DataFrame(df_genres.groupby(['genre'])['id_x'].count().sort_values(ascending=False)).reset_index().head(5)
df1
```

```python
sns.countplot(df_genres[df_genres['genre'].isin(['Thriller','Drama','Action','Adventure','Comedy'])]['genre'])
```

```python
#df1 = df1.reset_index()
genre_sales_by_year = df_genres.groupby(['year','genre'])['id_x'].count().reset_index().pivot_table(
                        index='year', 
                        columns='genre', 
                        values='id_x', 
                        aggfunc=sum).fillna(0).applymap(int)

```

```python
top_genres = pd.DataFrame(genre_sales_by_year[[x for x in genre_sales_by_year.columns if x != 'genre']].sum().sort_values(ascending = False))
```

```python
top_genres = top_genres.rename(index = str, columns = {0 : 'total_count'})
top_genres.head(5)
```

Отображаем общее количество релизов по жанрам с помощью гистограммы, переменные категориальные

```python

plt.barh(top_genres.reset_index()['genre'], top_genres.reset_index()['total_count'], color='green')

plt.ylabel("Genre")
plt.xlabel("Total released movies in this genre")
plt.title("The total count of genre released")
```

Рассмотрим отдельно ранее отобранные топ-5 жанров. Отобразим как менялось количество релизов в этих жанрах за определенные годы. Используем тип графика - bar, переменные категориальные

```python
genres = list(genre_sales_by_year.reset_index()['year'][genre_sales_by_year.reset_index()['year'].isin([1940,1960,1980,1990,2000,2005,2010,2015,2016])])
drama = np.array(genre_sales_by_year.reset_index()[genre_sales_by_year.reset_index()['year'].isin([1940,1960,1980,1990,2000,2005,2010,2015,2016])]['Drama'])
comedy = np.array(genre_sales_by_year.reset_index()[genre_sales_by_year.reset_index()['year'].isin([1940,1960,1980,1990,2000,2005,2010,2015,2016])]['Comedy'])
thriller = np.array(genre_sales_by_year.reset_index()[genre_sales_by_year.reset_index()['year'].isin([1940,1960,1980,1990,2000,2005,2010,2015,2016])]['Thriller'])
action = np.array(genre_sales_by_year.reset_index()[genre_sales_by_year.reset_index()['year'].isin([1940,1960,1980,1990,2000,2005,2010,2015,2016])]['Action'])
romance = np.array(genre_sales_by_year.reset_index()[genre_sales_by_year.reset_index()['year'].isin([1940,1960,1980,1990,2000,2005,2010,2015,2016])]['Romance'])
ind = [x for x, _ in enumerate(genres)]

plt.bar(ind, romance, width=0.8, label='romance', color='blue', bottom=drama+comedy+thriller+action)
plt.bar(ind, action, width=0.8, label='action', color='green', bottom=drama+comedy+thriller)
plt.bar(ind, thriller, width=0.8, label='thriller', color='gold', bottom=drama+comedy)
plt.bar(ind, comedy, width=0.8, label='comedy', color='silver', bottom=drama)
plt.bar(ind, drama, width=0.8, label='drama', color='#CD853F')

plt.xticks(ind, genres)
plt.ylabel("Total released")
plt.xlabel("Years")
plt.legend(loc="upper left")
plt.title("Genres released by years")

plt.show()

```
