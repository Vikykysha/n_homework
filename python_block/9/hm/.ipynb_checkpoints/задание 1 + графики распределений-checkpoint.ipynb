```python
def dont_beat(n_lad, frame_size):
    #Количество способов расставить n_lad ладей на доске размера frame_size
    cnt_comb = comb(frame_size ** 2, n_lad, exact = True)
    #Количество способов расставить ладей, чтобы они не били друг друга. Закрепляем вертикали и для каждой ищем способ поставить на нее очередную ладью
    cnt_dont_beat = perm(frame_size,frame_size)
    return cnt_dont_beat / cnt_comb
```

```python
dont_beat(4,4)
```

```python
dont_beat(8,8)
```

Посмотрим на распределение вероятности в зависимости от числа n.

```python
trials = np.array([])
```

```python
for x in range(1,11):
    trials = np.append(trials,dont_beat(x,x))
trials
```

```python
#зададим функцию для построения cdf
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1,n+1)/n
    return x,y
```

```python
#получим значения для наших trials
x, y = ecdf(trials)
```

```python
#отобразим ecdf на графике
plt.plot(x,y,marker = '.', linestyle = 'none');
```

Пункт b

Обозначим следующие события
A = {ладьи не бьют друг друга}
B = {ни одна из ладей не стоит на главной диагонали}
Di = {i-я клетка главной диагонали занята ладьей и ни одна не бьет другую}


Тогда \\( \cup Di \\) даст нам событие \\( \neg(A \cap B)  \\) - {хотя бы одна ладья стоит на главной диагонали  и ни одна ладья не бьет другую}


Cобытие \\( A \cap B \\) - {ни одна ладья не стоит на главной диагонали и ни одна ладья не бьет другую}

Тогда \\( \neg(A \cap B) \cup (A \cap B) = A \\)

Тогда искомое событие будет равняться \\( A - \neg(A \cap B) \\)

Число исходов события A = n\\(!\\)

Число \\( \neg(A \cap B) \\) можно получить используя формулу включений/исключений

```python
def count_subfactorial(n, start = 2, sign = -1):
    """Подсчитывает субфакториал от n"""
    if abs(start) > n:
        return 0
    sign = -1 * sign
    return sign * 1/perm(start, start) +  count_subfactorial(n, start=start + 1, sign = sign)
```

```python
def count_dont_beat_diag_out(n):
    """Возвращает вероятность расставить n ладей так, чтобы ни одна не била другую и на главной диагонали не было фигур"""
    return (perm(n,n) * count_subfactorial(n))/comb(n**2,n)
```

```python
count_dont_beat_diag_out(3)
```

```python
trials = np.array([])
for x in range(1,11):
    trials = np.append(trials,count_dont_beat_diag_out(x))
trials
```

```python
x, y = ecdf(trials)
```

```python
#отобразим ecdf на графике
plt.plot(x,y,marker = '.', linestyle = 'none');
```

#### Пуассон

```python
np.random.seed(42)
```

```python
samples = np.random.poisson(6, 10000)
```

```python
x,y = ecdf(samples)
```

```python
plt.plot(x,y,marker='.',linestyle='none')
plt.margins(0.02)
plt.xlabel('numbers of success')
plt.ylabel('ECDF');
```

```python
bins = np.arange(0, max(samples) + 1.5) - 0.5
```

```python
plt.hist(samples, normed = True, bins=bins)
plt.xlabel('number of success')
plt.ylabel('PMF');
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 15, 10
size = [100, 500, 1000]
cnt = [6, 9, 11]
prob = [0.5, 0.1, 0.01]
```

```python
for i in range(0,3):
    samples_binomial = np.random.binomial(size[i],prob[i], size = 10000)
    samples_p = np.random.poisson(cnt[i], 10000)
    plt.subplot(3,1,i+1)
    x_b, y_b = ecdf(samples_binomial)
    x_p, y_p = ecdf(samples_p)
    plt.plot(x_b, y_b, marker = '.', linestyle = 'none',label = 'binomial')
    plt.plot(x_p, y_p, marker = '.', linestyle = 'none', label = 'poisson')
    plt.xlabel('number of success')
    plt.ylabel('ECDF')
plt.legend(loc = 'right');
```

По графикам выше можно видеть, что при большом количестве испытаний и при маленькой верояности биномиальное распределение приближается к распределению Пуассона. При обратных условиях, разница между ними достаточно велика.

Построим pmf для распределения Пуассона, используя модуль scipy.stats

```python
from scipy.stats import poisson
```

```python
x = np.arange(0,200,0.5)
for i in [50,80,120]:
    plt.plot(x, poisson.pmf(x,i), label = "N = 200, mean = %s" % (str(i)))
plt.xlabel('number of success')
plt.ylabel('probability')
plt.legend(loc='upper right')
plt.title('Poisson PMF');
```

По графикам можно видеть, что при повышении среднего, разброс значений повышается при неизменном количестве испытаний.

#### Экспоненциальное распределение

```python
from scipy.stats import expon
```

```python
rv = expon(scale = 1/6)
x = np.linspace(0,4,100)
plt.plot(x,rv.pdf(x))
plt.xlabel()
```

```python
plt.plot(x,rv.pdf(x))
plt.xlabel("time between events")
plt.ylabel('PDF');
```

```python
rate = [1/6,1/12,1]
x=np.linspace(0,5,100)
for i in range(3):
    plt.plot(x,expon.pdf(x, scale=rate[i]), label = "Rate = %s" % (str(round(rate[i],3))))
    plt.xlabel("Time between events")
    plt.ylabel("PDF")
plt.legend();
```

```python
rate = [1/6,1/12,1]
x=np.linspace(0,5,100)
for i in range(3):
    plt.plot(x,expon.cdf(x, scale=rate[i]), label = "Rate = %s" % (str(round(rate[i],3))))
    plt.xlabel("Time between events")
    plt.ylabel("CDF")
plt.legend();
```

#### Геометрическое распределение

```python
from scipy.stats import geom
```

```python
plt.figure(figsize=(10,7)); plt.subplot(2,1,1);

prop = 0.25;
ulim = int(4/prop); # choose 4*mean as the upper limit of the support interval
# first plot the PMF
x = np.arange(1,ulim); # for geometric variables, start at one (i.e., one trial)
p = geom.pmf(x,prop); # compute the PMF (the order of arguments is n,p)
plt.stem(x,p); # the stem plot is most appropriate for plotting PMFs; it illustrates that probability is only defined at integer values
plt.axis([0.5,ulim+0.5,0,0.3]) # adjust the axis limits
plt.title('PMF');  plt.ylabel(r'$P(x)$');

# now plot the CDF
plt.subplot(2,1,2);
P = geom.cdf(x,prop); # compute the CDF
plt.step(x,P); # the step plot is most appropriate for plotting discrete CDFs;
plt.axis([0.5,ulim+0.5,0,1.01]); # adjust the axis limits
plt.title('CDF');  plt.ylabel(r'$P(X\leq x)$');
```
