Можем видеть, что признаки area_mean, area_worst, area_se имеют достаточно много выбросо в данных. Избавимся от них, так как они могут повлиять на качество модели, кластеризации.

```python
Q1 = df_raw[['area_mean','area_se','area_worst']].quantile(0.25)
Q3 = df_raw[['area_mean','area_se','area_worst']].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df_raw_out = df_raw[~((df_raw[['area_mean','area_se','area_worst']] < (Q1 - 1.5 * IQR)) |(df_raw[['area_mean','area_se','area_worst']] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df_raw_out.shape
```

```python
fig,axes = plt.subplots(15,2,figsize=(10,20))
malignant = df_raw.drop('diagnosis',axis=1)[df_raw['diagnosis_c']==1]
benign = df_raw.drop('diagnosis',axis=1)[df_raw['diagnosis_c']==0]

ax = axes.ravel()

for i in range(30):
    _, bins = np.histogram(df_raw.drop('diagnosis',axis=1).iloc[:,i],bins=50)
    ax[i].hist(malignant.iloc[:,i],bins=bins,color='b',alpha=.5)
    ax[i].hist(benign.iloc[:,i],bins=bins,color='g',alpha=.5)
    ax[i].set_title(benign.iloc[:,i].name)
    ax[i].set_yticks(())
ax[0].set_xlabel("Значения признака")
ax[0].set_ylabel("Частота")
ax[0].legend(['доброкачественная','злокачественная'],loc='best')
fig.tight_layout()
```

Из-за большого количества признаков, достаточно трудно уловить какую-то информацию из графиков выше. Однако, мы теперь можем примерно понимать, какие признаки достаточно хорошо должны объяснять зависимую переменную. Например, признак smoothness_se похоже, малоинформативен, так как обе гистограммы практически накладываются друг на друга, в то время как признак concave points_worst информативен, так как две гистограммы практически не накрывают друг друга. Также, достаточно информативным является признаки radius_worst,concave points_mean,concativity_worst,perimeter_worst.

Давайте проверим, является ли статистически значимым различие между классами для некоторых признаков

```python
d_m = df_raw[df_raw['diagnosis_c']==1]
d_b = df_raw[df_raw['diagnosis_c']==0]
```

```python
sts.ttest_ind(d_m['concave points_worst'], d_b['concave points_worst'])
```

При данном уровне значимости (0.05) мы можем отклонить нулевую гипотезу и сказать, что существует статистически значимое различие между средним значением 'concave points_worst' для данных двух классов. 

```python
sts.ttest_ind(d_m['concavity_worst'], d_b['concavity_worst'])
```

При данном уровне значимости (0.05) мы можем отклонить нулевую гипотезу и сказать, что существует статистически значимое различие между средним значением 'concavity_worst' для данных двух классов. 

```python
sts.ttest_ind(d_m['radius_se'], d_b['radius_se'])
```

При данном уровне значимости (0.05) мы можем отклонить нулевую гипотезу и сказать, что существует статистически значимое различие между средним значением 'concavity_worst' для данных двух классов. 

```python
sts.ttest_ind(d_m['perimeter_worst'], d_b['perimeter_worst'])
```

При данном уровне значимости (0.05) мы можем отклонить нулевую гипотезу и сказать, что существует статистически значимое различие между средним значением 'concavity_worst' для данных двух классов. 

Таким образом, наши предположения, сделанные по гистограммам, подтверждаются, те признаки, гистограммы которых сильно не пересекаются, действительно имеют значимое различие внутри рассматривамых классов.

Попробуем визуализировать наши данные используя PCA, выдели только две первые компоненты

Так как многие из признаков имеют разный масштаб, предварительно отмасштабируем данные

```python
scaler = StandardScaler()
scaler.fit(df_raw_out.drop('diagnosis',axis=1).values)
X_scaled = scaler.transform(df_raw_out.drop('diagnosis',axis=1).values)
```

```python
pca = PCA(n_components=2)
pca.fit(X_scaled)
pca.transform(X_scaled)
X_pca = pca.transform(X_scaled)
print('Формат исходного массива: {}'.format(str(X_scaled.shape)))
print('Формат массива после сокращения размерности: {}'.format(str(X_pca.shape)))
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X_pca[:,0][df_raw_out['diagnosis_c']==0],X_pca[:,1][df_raw_out['diagnosis_c']==0],c='g',alpha=.8)
plt.scatter(X_pca[:,0][df_raw_out['diagnosis_c']==1],X_pca[:,1][df_raw_out['diagnosis_c']==1],c='b',alpha=.8)
plt.gca().set_aspect('equal')
plt.legend(['benign','malignant'],loc='best')
plt.xlabel("Первая главная компонента")
plt.ylabel("Вторая главная компонента");
```

Можем видеть, что в  двумерном пространстве два класса разделены достаточно хорошо, что говорит нам о том, что даже линейные методы машинного обучения позволят нам добиться неплохих результатов при классфикации.

Попробуем визуализировать наши данные , разбив их на кластеры

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X_scaled)
```

```python
kmeans.labels_
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X_pca[:,0][kmeans.labels_==0],X_pca[:,1][kmeans.labels_==0],c='g',alpha=.8)
plt.scatter(X_pca[:,0][kmeans.labels_==1],X_pca[:,1][kmeans.labels_==1],c='b',alpha=.8)
plt.gca().set_aspect('equal')
plt.legend(['benign','malignant'],loc='best');
```

Как мы можем видеть, алгоритм k-средних достаточно хорошо справился с задачей разделения данных на классы. Опять можно проследить, что классы достаточно хорошо можно разделить прямой линией.

Можем видеть, что наше разбиение на кластеры имеет достаточну высокую инентность

```python
print(kmeans.inertia_)
```

```python
ks = range(1, 8)
inertias = []

for k in ks:

    model = KMeans(n_clusters = k)

    model.fit(X_scaled)

    inertias.append(model.inertia_ )
    
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

Если посмотреть на график и попытаться применить правило локтя, то оптимальным кажется вариант из 3 кластеров

```python
kmeans_3 = KMeans(n_clusters=3)
kmeans_3.fit(X_scaled)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X_pca[:,0][kmeans_3.labels_==0],X_pca[:,1][kmeans_3.labels_==0],c='g',alpha=.8)
plt.scatter(X_pca[:,0][kmeans_3.labels_==1],X_pca[:,1][kmeans_3.labels_==1],c='b',alpha=.8)
plt.scatter(X_pca[:,0][kmeans_3.labels_==2],X_pca[:,1][kmeans_3.labels_==2],c='y',alpha=.8)
plt.gca().set_aspect('equal')
plt.legend(['benign','malignant','other'],loc='best');
```

По графику можем видеть, что, похоже, что в третий кластер попали outliers из рассматриваемых классов. Мы удалили outliers только у некоторых признаков, видимо, у других признаков выбросы тоже остались, и, как можно видеть, оказываются влияние. Если бы мы не знали, что на самом деле класса у нас два, то могли бы принять данные из зашумленного третьего кластера за отдельный , полноценные кластер, что могло привести к неверным интерпретациям данных.

Попробуем построить baseline модель и сравнить ее результаты в следующих ситуациях:

2. Воспользумеся методом отбора признаков на основе модели

###### 5.Baseline model

```python
skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
logit = LogisticRegression(n_jobs=-1,random_state=42)
```

```python
X = df_raw_out.drop(['diagnosis_c','diagnosis'],axis=1)
y = df_raw_out['diagnosis_c']
```

```python
scores = cross_validate(
    logit, X, y, cv=skf, scoring=('roc_auc', 'accuracy', 'f1', 'precision'), return_train_score=True
)
```

```python
scores
```

```python
scores1 = scores['train_roc_auc']
scores1
```

```python
scores1.mean()
```

Можем видеть, что уже простая логистическая регрессия на первоначальных данных дает высокие результаты

##### 1.PCA

```python
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
pca = PCA().fit(X_scaled)
x_pca = pca.transform(X_scaled)
```

```python
cum_evr = np.cumsum(pca.explained_variance_ratio_)
xs = np.arange(cum_evr.size) + 1
plt.plot(xs, cum_evr, linewidth=2)
plt.grid()
plt.xlabel('n_components')
plt.ylabel('explained_variance_ratio_')
```

Можем видеть, что уже 4 компоненты объясняют порядка 80% дисперсии в данных. 25 компонент объясняют 100% дисперсии.

```python
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
pca = PCA(n_components=25).fit(X_scaled)
x_pca = pca.transform(X_scaled)
```

```python
scores2 = cross_val_score(
    logit, x_pca, y, cv=skf, scoring='roc_auc'
)
```

```python
scores2 
```

```python
scores2  > scores1
```

```python
scores2.mean()
```

Используя только 25 первых компонент, мы получили хушие результаты

## 

#### 3.Одномерный способ отбора признаков

```python
from sklearn.feature_selection import SelectPercentile
```

```python
select = SelectPercentile(percentile=80)
select.fit(X_scaled,y)
X_selected = select.transform(X_scaled)
```

```python
scores3 = cross_val_score(
    logit, X_selected, y, cv=skf, scoring='roc_auc'
)
```

```python
scores3
```

```python
scores3 > scores2
```

```python
scores3.mean()
```

```python
X_selected.shape
```

Используя только 80% отобранных признаков мы получили оценки выше, чем у baseline модели

#### 2.Воспользумеся методом отбора признаков на основе модели

Для отбора признаков на основе модели, используем модель деревья решений

```python
from sklearn.feature_selection import SelectFromModel
```

```python
select = SelectFromModel(RandomForestClassifier(n_estimators=100,random_state=42),threshold=1e-5
                        )
```

```python
select.fit(X_scaled,y)
X_selected_tree = select.transform(X_scaled)
```

```python
scores4 = cross_val_score(
    logit, X_selected_tree, y, cv=skf, scoring='roc_auc'
)
```

```python
scores4.mean()
```

```python
scores4 > scores3
```

Данный способ, дал худшие результаты, по сравнению со способом одномерного отбора признаков.

#### 4.Воспользуемся итеративным способом отбора признаков

```python
from sklearn.feature_selection import RFE
```

```python
select = RFE(RandomForestClassifier(n_estimators=100, random_state=42),n_features_to_select=26)
select.fit(X_scaled,y)
```

```python
X_rfe = select.transform(X_scaled)
```

```python
scores5 = cross_val_score(
    logit, X_rfe, y, cv=skf, scoring='roc_auc'
)
```

```python
scores5.mean()
```

```python
scores5> scores4
```

```python
scores5> scores1
```

```python
scores5> scores4
```

RFE дал результат лучше , чем отбор на основе модели

```python
x = np.array([scores1.mean(),scores2.mean(),scores3.mean(),scores4.mean(),scores5.mean()])
x.argmax()
```

Можем видеть, что самые высокие оценки получил итеративный отбор признаков
